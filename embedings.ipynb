```python
import json
from openai import OpenAI
from config import OPENAI_API_KEY
from data_loader import load_data_from_file
```

```python
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
# Load FAQ files
with open("data/faq_georgian.json", "r", encoding="utf-8") as f:
    faq_georgian = json.load(f)
```

```python
with open("data/faq_english.json", "r", encoding="utf-8") as f:
    faq_english = json.load(f)
    
```

```python
with open("data/intents.json", "r", encoding="utf-8") as f:
    intents = json.load(f)
```

```python
# Function to generate embeddings
def generate_embeddings(faq_dict):
    embeddings = {}
    for question in faq_dict.keys():
        response = client.embeddings.create(
            model="text-embedding-3-small",
            input=question
        )
        embeddings[question] = response.data[0].embedding
    return embeddings
```

```python
# Generate and save embeddings
faq_embeddings_georgian = generate_embeddings(faq_georgian)
faq_embeddings_english = generate_embeddings(faq_english)
intents_embeddings = generate_embeddings(intents)
```

```python
with open("data/faq_georgian_embeddings.json", "w", encoding="utf-8") as f:
    json.dump(faq_embeddings_georgian, f)

with open("data/faq_english_embeddings.json", "w", encoding="utf-8") as f:
    json.dump(faq_embeddings_english, f)
    
with open("data/intents_embeddings.json", "w", encoding="utf-8") as f:
    json.dump(intents_embeddings, f)
```
